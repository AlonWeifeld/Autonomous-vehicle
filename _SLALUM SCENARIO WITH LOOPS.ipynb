# TELE

```python
Slalom = pd.read_csv("D:\\Alon\\calc_events_new_up.csv")
Slalom = Slalom[Slalom.Scenario == 'Slalom']
#display(Slalom)
event_to_check = [('Start teleoperation session 1', 'End teleoperation session 1'),
                  ('Start teleoperation session 2', 'End teleoperation session 2') ]
conditions = Slalom.Condition.unique()
ids = Slalom.Id.unique()
time_of_event     = []
dist_of_event     = []
max_braking       = []
v = 0
c = 0

for _id in ids:
    tmp_Slalom_id = Slalom[Slalom.Id == _id]

    for condition in conditions:
        tmp_Slalom_condition = tmp_Slalom_id[tmp_Slalom_id.Condition == condition]    
    
        for event in event_to_check:
            tmp_Slalom_event = tmp_Slalom_condition[tmp_Slalom_condition['loc'] == event[0]]
            
            try:
                t = 0
                if event[0] == event[1]:
                    t += 1
                    time_start = tmp_Slalom_event.iloc[0]['SimulationTime']
                    dist_start = tmp_Slalom_event.iloc[0]['Distance_Driven']
                    t += 1 #1 
                    time_end   = tmp_Slalom_event.tail(1)['SimulationTime'].values[0]
                    dist_end   = tmp_Slalom_event.tail(1)['Distance_Driven'].values[0]
                    
                else:
                    t += 1
                    time_start = tmp_Slalom_event.iloc[0]['SimulationTime']
                    dist_start = tmp_Slalom_event.iloc[0]['Distance_Driven']
                    t += 1 #1
                    time_end   = tmp_Slalom_condition[tmp_Slalom_condition['loc'] == event[1]].iloc[0]['SimulationTime']
                    dist_end   = tmp_Slalom_condition[tmp_Slalom_condition['loc'] == event[1]].iloc[0]['Distance_Driven']
                    t += 1 #2

                delta_time = time_end - time_start
                if c < 1:
                    print(time_start, time_end, delta_time)
                    c += 1
                    
                delta_dist = dist_end - dist_start
                t += 1 #3
                event_id = tmp_Slalom_event.iloc[0]['Event_Name']
                row      = tmp_Slalom_event[tmp_Slalom_event.Event_Name == 'Braking']
                #display(row)
                #display(row)
                t += 1 #4

                time_of_event.append((_id, condition, event_id, time_start, time_end, delta_time))
                dist_of_event.append((_id, condition, event_id, dist_start, dist_end, delta_dist))
                                
                row  = row.loc[row['Amplitude'].idxmax()]
                
                amp  = row['Amplitude']
                
                time_amp   = row['SimulationTime'      ]
                time_onset = row['Onset.SimulationTime']
                time_end   = row['End.SimulationTime'  ]
                
                dist_amp   = row['Distance_Driven'      ]
                dist_onset = row['Onset.Distance_Driven']
                dist_end   = row['End.Distance_Driven'  ]
                
                max_braking.append((_id, condition, event_id, amp, time_onset, time_amp, time_end, dist_onset, dist_amp, dist_end))
                
            except: # catch *all* exceptions
                e = sys.exc_info()
                v += 1
                print(t, e, _id, condition, event[0])
                #print('*********')
                #print('')
                #print('*********')
```

```python
df_times_tele       = pd.DataFrame(time_of_event, columns = ['Id', 'Condition', 'Event_Name', 'Time_Start', 'Time_End', 'Total_Time'])
df_dists_tele       = pd.DataFrame(dist_of_event, columns = ['Id', 'Condition', 'Event_Name', 'Distance_Start', 'Distance_End', 'Total_Distance'])
df_max_braking_tele = pd.DataFrame(max_braking,   columns = ['Id', 'Condition', 'Event_Name', 'Amplitude', 'Onset.SimulationTime', 'SimulationTime', 'End.SimulationTime', 'Onset.Distance_Driven', 'Distance_Driven', 'End.Distance_Driven'])


display(df_times_tele)
display(df_dists_tele)
display(df_max_braking_tele)
```

```python
df_times_tele       = pd.DataFrame(time_of_event, columns = ['Id', 'Condition', 'Event_Name', 'Time_Start', 'Time_End', 'Total_Time'])

df_times_tele['Condition'] = df_times_tele['Condition'].str.replace('Slalom1_close', '1')
df_times_tele['Condition'] = df_times_tele['Condition'].str.replace('Slalom1_far', '1')
df_times_tele['Condition'] = df_times_tele['Condition'].str.replace('Slalom2_far', '2')
df_times_tele['Condition'] = df_times_tele['Condition'].str.replace('Slalom2_close', '2')
df_times_tele
```

```python
df_times_tele1      = pd.DataFrame(df_times_tele, columns = ['Id','Condition','Event_Name', 'Total_Time'])
compare_time= df_times_tele1.groupby(['Id','Condition'])[['Total_Time']].sum()


df_latecny1= df_times_tele[df_times_tele.Condition=="1"]

df_latency2= df_times_tele[df_times_tele.Condition=="2"]

df1_merge = df_latecny1.merge(df_latency2, how = 'inner', on=['Id','Event_Name'])
df_times_tele_merge      = pd.DataFrame(df1_merge, columns = ['Id','Event_Name','Total_Time_x', 'Total_Time_y'])
df_times_tele_merge['ratio_tele'] = df_times_tele_merge.Total_Time_x/df_times_tele_merge.Total_Time_y
df_times_tele_merge['Event_Name'] = df_times_tele['Event_Name'].str.replace('teleoperation', '')
df_times_tele_merge.rename({'Total_Time_x': 'Total_time_latency1',
                            'Total_Time_y': 'Total_time_latency2'}, axis=1, inplace=True)
df_times_tele_merge
#compare_time
#df_times_tele.Condition
#SPLIT CONDITION 1 AND 2 TO SPERAD TABELS
#total_time_x=total_time_latency1
#total_time_y=total_time_latency2
```
